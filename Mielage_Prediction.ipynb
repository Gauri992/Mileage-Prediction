**Mileage Prediction**


**Objective**

This project is about predicting the mileage of a particular car in city driving, given data of some parameters (features) for cars.

**Data Source**


https://github.com/YBI-Foundation/Dataset/blob/main/MPG.csv

6) acceleration:  continuous

8) origin:        multi-valued discrete

**Import Library**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Import Data**

```python
df = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/MPG.csv')
```

**Describe Data**

```python
df.head()
```

```python
df.nunique()
```

```python
df.info()
```

```python
df.corr()
```

**Data Preprocessing**

```python
df = df.dropna()
```

```python
df.info()
```

**Data Visualization**

```python
sns.pairplot(df,x_vars = ['displacement','horsepower','weight','acceleration','mpg'],y_vars=['mpg']);
```

```python
sns.regplot(x = 'displacement',y = 'mpg' , data = df);
```

```python
df.columns
```

**Define Target variable(y) and feature variables (X)**

```python
y = df['mpg']
```

```python
y.shape
```

```python
X = df[['displacement','horsepower','weight','acceleration']]
```

```python
X.shape
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
X = ss.fit_transform(X)
```

```python
X
```

```python
pd.DataFrame(X).describe()
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, train_size =0.7 ,random_state = 2529)
```

```python
X_train.shape ,X_test.shape ,y_train.shape , y_test.shape
```

**Modeling**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train ,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

**Prediction**

Mileage = 23.4 -1.05Displacement -1.68Horsepower - 4.10weight -0.115Acceleration + error

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

**Model Evaluation**

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,r2_score
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test ,y_pred)
```

```python
r2_score(y_test , y_pred)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree = 2 ,interaction_only = True ,include_bias = False)
```

```python
X_train2 = poly.fit_transform(X_train)
```

```python
X_test2 = poly.fit_transform(X_test)
```

```python
lr.fit(X_train2,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_pred_poly = lr.predict(X_test2)
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,r2_score
```

```python
mean_absolute_error(y_test ,y_pred_poly)
```

```python
mean_absolute_percentage_error(y_test ,y_pred_poly)
```

```python
r2_score(y_test ,y_pred_poly)
```
